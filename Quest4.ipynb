# 1. 손글씨 분류

```python
# 필요한 모듈 import하기 
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# 데이터 준비 
num_digits = load_digits()
num_digits.keys()
```

```python
# 데이터 이해하기 
num_data = num_digits.data
label_data = num_digits.target
num_data.shape
```

```python
label_data
```

```python
num_digits.feature_names
```

```python
num_digits.target_names
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for i in range(30):
    plt.subplot(6, 5, i+1)
    plt.imshow(num_data[i].reshape(8,8), cmap ='gray')
    plt.axis('off')
```

```python
print(num_digits.DESCR)
```

```python
# train, test 데이터 분리
import matplotlib.pyplot as plt
X_train, X_test, y_train, y_test = train_test_split(num_data, label_data, test_size = 0.2, random_state = 32)
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model_tree = DecisionTreeClassifier(random_state = 32)
model_tree.fit(X_train, y_train)
tree_pred = model_tree.predict(X_test)

# Random Forest
from sklearn.ensemble import RandomForestClassifier

model_random = RandomForestClassifier(random_state = 32)
model_random.fit(X_train,y_train)
random_pred = model_random.predict(X_test)

# SVM
from sklearn import svm

model_svm = svm.SVC(random_state = 32)
model_svm.fit(X_train, y_train)
svm_pred = model_svm.predict(X_test)

# SGD classifier
from sklearn.linear_model import SGDClassifier

model_sgd = SGDClassifier(random_state = 32)
model_sgd.fit(X_train, y_train)
sgd_pred = model_sgd.predict(X_test)

# Logistic Regression 
from sklearn.linear_model import LogisticRegression

model_logi = LogisticRegression(random_state = 32, max_iter = 3000)
model_logi.fit(X_train, y_train)
logi_pred = model_logi.predict(X_test)

```

```python
# 모델을 평가해보기 

# 정확도 
logi_accuracy = accuracy_score(y_test, logi_pred)
sgd_accuracy = accuracy_score(y_test, sgd_pred)
svm_accuracy = accuracy_score(y_test, svm_pred)
random_accuracy = accuracy_score(y_test, random_pred)
tree_accuracy = accuracy_score(y_test, tree_pred)

print(f'Decission Tree accuracy : {tree_accuracy} \nRandom Forest : {random_accuracy} \nSVM accuracy : {svm_accuracy} \nSGD accuracy : {sgd_accuracy} \nLogistic Regression accuracy : {logi_accuracy}')
      
```

```python
# 결과지표 확인
print(f'Decission Tree : \n {classification_report(y_test, tree_pred)}')
print()
print(f'Random Forest : \n {classification_report(y_test, random_pred)}')
print()
print(f'SVM : \n {classification_report(y_test, svm_pred)}')
print()
print(f'SGD : \n {classification_report(y_test, sgd_pred)}')
print()
print(f'Logistic Regression :\n {classification_report(y_test, logi_pred)}')
```

# 2. 와인 분류

```python
# 모듈 import하기 
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# 데이터 준비하기

wine_data = load_wine()
wine_data.keys()
```

```python
print(len(wine_data.data))
wine_data.data
```

```python
print(len(wine_data.target))
wine_data.target
```

```python
# 데이터 이해하기

wine_feature = wine_data.data
wine_target  = wine_data.target
wine_data.target_names
```

```python
print(wine_data.DESCR)
```

```python
wine_feature.shape
```

```python
import pandas as pd
wine_df = pd.DataFrame(data = wine_feature, columns = wine_data.feature_names)
wine_df['class'] = wine_data.target
wine_df.head()
```

```python
# train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(wine_feature, wine_target, test_size = 0.2, random_state = 32)
```

```python
# 다양한 모델로 학습시키기

# Decision Tree

wine_tree_model = DecisionTreeClassifier(random_state = 32)
wine_tree_model.fit(X_train, y_train)
wine_tree_pred = wine_tree_model.predict(X_test)

# Random Forest

wine_rand_model = RandomForestClassifier(random_state = 32)
wine_rand_model.fit(X_train, y_train)
wine_rand_pred = wine_rand_model.predict(X_test)

# SVM

wine_svm_model = svm.SVC(random_state = 14)
wine_svm_model.fit(X_train, y_train)
wine_svm_pred = wine_svm_model.predict(X_test)

#SGD

wine_sgd_model = SGDClassifier(random_state = 14)
wine_sgd_model.fit(X_train, y_train)
wine_sgd_pred = wine_sgd_model.predict(X_test)

# Logistic Regression

wine_logi_model = LogisticRegression(random_state = 32, max_iter = 10000)
wine_logi_model.fit(X_train, y_train)
wine_logi_pred = wine_logi_model.predict(X_test)
```

```python
# 모델을 평가해 보기

# 정확도
wine_tree_accu = accuracy_score(y_test, wine_tree_pred)
wine_rand_accu = accuracy_score(y_test, wine_rand_pred)
wine_svm_accu = accuracy_score(y_test, wine_svm_pred)
wine_sgd_accu = accuracy_score(y_test, wine_sgd_pred)
wine_logi_accu = accuracy_score(y_test, wine_logi_pred)

print(f'Decission Tree accuracy : {wine_tree_accu} \nRandom Forest : {wine_rand_accu} \nSVM accuracy : {wine_svm_accu} \nSGD accuracy : {wine_sgd_accu} \nLogistic Regression accuracy : {wine_logi_accu }')
 
```

```python
# 결과지표 확인
print(f'Decission Tree : \n {classification_report(y_test, wine_tree_pred)}')
print()
print(f'Random Forest : \n {classification_report(y_test, wine_rand_pred)}')
print()
print(f'SVM : \n {classification_report(y_test, wine_svm_pred)}')
print()
print(f'SGD : \n {classification_report(y_test, wine_sgd_pred)}')
print()
print(f'Logistic Regression :\n {classification_report(y_test, wine_logi_pred)}')
```

```python
# 필요한 모듈 import

from sklearn.datasets import load_breast_cancer
```

```python
breast_feature = breast_data.data
breast_target = breast_data.target
print(len(breast_feature))
print(len(breast_target))
```

```python
# 데이터 이해하기
print(breast_data.DESCR)
```

```python
breast_cancer = pd.DataFrame(data = breast_feature, columns = breast_data.feature_names)
breast_cancer
```

```python
# train, test 데이터 분리

X_train, X_test, y_train, y_test = train_test_split(breast_feature, breast_target, test_size = 0.2, random_state = 32)
print(len(X_train))
print(len(y_train))
```

```python
# 다양한 모데로 학습시켜보기

# Decision Tree 
breast_tree_model = DecisionTreeClassifier(random_state = 32)
breast_tree_model.fit(X_train, y_train)
breast_tree_pred = breast_tree_model.predict(X_test)

# Random Forest
breast_rand_model = RandomForestClassifier(random_state = 32)
breast_rand_model.fit(X_train, y_train)
breast_rand_pred = breast_rand_model.predict(X_test)

# SVM
breast_svm_model = svm.SVC(random_state = 32)
breast_svm_model.fit(X_train, y_train)
breast_svm_pred = breast_svm_model.predict(X_test)

# SGD Classifier 
breast_sgd_model = SGDClassifier(random_state = 32)
breast_sgd_model.fit(X_train, y_train)
breast_sgd_pred = breast_sgd_model.predict(X_test)

# Logistic Regression
breast_logi_model = LogisticRegression(random_state = 32, max_iter = 3000)
breast_logi_model.fit(X_train, y_train)
breast_logi_pred = breast_logi_model.predict(X_test)
```

```python
# 모델을 평가해 보기

# 정확도
wine_tree_accu = accuracy_score(y_test, breast_tree_pred)
wine_rand_accu = accuracy_score(y_test, breast_rand_pred)
wine_svm_accu = accuracy_score(y_test, breast_svm_pred)
wine_sgd_accu = accuracy_score(y_test, breast_sgd_pred)
wine_logi_accu = accuracy_score(y_test, breast_logi_pred)

print(f'Decission Tree accuracy : {wine_tree_accu} \nRandom Forest : {wine_rand_accu} \nSVM accuracy : {wine_svm_accu} \nSGD accuracy : {wine_sgd_accu} \nLogistic Regression accuracy : {wine_logi_accu }')
 
```

```python
# 결과지표 확인
print(f'Decission Tree : \n {classification_report(y_test, breast_tree_pred)}')
print()
print(f'Random Forest : \n {classification_report(y_test, breast_rand_pred)}')
print()
print(f'SVM : \n {classification_report(y_test, breast_svm_pred)}')
print()
print(f'SGD : \n {classification_report(y_test, breast_sgd_pred)}')
print()
print(f'Logistic Regression :\n {classification_report(y_test, breast_logi_pred)}')
```
