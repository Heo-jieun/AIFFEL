### 라이브러리 불러오기

```python
import os
import requests
from io import BytesIO
from itertools import product
from datetime import datetime

# 데이터 처리
import pandas as pd
import numpy as np

# 시각화
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
import seaborn as sns

# 분석
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV
import statsmodels.api as sm
import pmdarima as pm  

# statsmodels
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import STL
from statsmodels.tsa.stattools import adfuller # 정상성 검정
from statsmodels.tsa.arima_model import ARIMA

import warnings
warnings.filterwarnings( 'ignore' )

print(pd.__version__)
print(sm.__version__)
print(pm.__version__)
```

### Step 1. 학습 데이터와 테스트 데이터 나누기

```python
# 파일 불러오기
file_path_raws = os.getenv('HOME') + '/aiffel/demand_forecasting/data/raws.csv'
df_raws = pd.read_csv(file_path_raws)

# 설명 변수 테이블 설정
_x_data = df_raws[['click_d_1', 'click_d_2', 'click_d_3', 'click_d_4', 'click_d_5', 'click_d_6', 'click_d_7', 'is_clean', 'avg_precipitation', 'avg_temperature', 'target_dates']]
_x_weekday = pd.get_dummies(df_raws['weekday'], prefix='weekday', drop_first=False)
_x_data = pd.concat([_x_data, _x_weekday], axis=1)
_x_data.set_index("target_dates", drop=True, inplace=True)
_x_data = _x_data.sort_values('target_dates')

# 예측 대상 데이터 설정
target = df_raws[['op_rate_0d_all_cars', 'target_dates' ]]
target.set_index("target_dates", drop=True, inplace=True)
target = target.sort_values('target_dates')

# Train/Test 데이터셋 나누기
_x_data_train = _x_data[_x_data.index<'2021-04-01']
_x_data_test = _x_data[_x_data.index>='2021-04-01']
target_train = target[target.index<'2021-04-01']
target_test = target[target.index>='2021-04-01']

x_train = _x_data_train
x_test = _x_data_test
y_train = target_train
y_test = target_test
```

```python
print(len(_x_data))
```

### Step 2. Grid search를 이용하여 최상의 파라미터 조합 찾기

```python
# 모수
p = range(0, 2)
q = range(0, 2)
d = range(0, 2)
P = range(0, 2)
Q = range(0, 2)
D = range(0, 2)
m = 7
# 초기화
best_AIC = float("inf")
best_model = sm.tsa.SARIMAX(endog=y_train, exog=x_train, order=(0, 0, 0), seasonal_order=(0, 0, 0, 7))
iter_cnt = 0
trend_pdq = list(product(p,d,q))
seasonal_pdq = [(val[0], val[1], val[2], m) for val in list(product(P,D,Q))]

# for 문으로 최소 AIC 찾기 (Grid Search)
AIC =  []
SARIMAX_order = []

for trend_param in trend_pdq:
    for seasonal_param in seasonal_pdq :
        try : 
    
            iter_cnt += 1
            mod = sm.tsa.SARIMAX(endog = y_train, exog = x_train, 
                                        order = trend_param, seasonal_order = seasonal_param)
            mod_fit = mod.fit(disp = False, maxiter = 250)
            print('Iter {}/64 : SARIMAX({})x({})({}), AIC ={}'.format(iter_cnt,trend_param,seasonal_param[:3],seasonal_param[3],mod_fit.aic))
            AIC.append(mod_fit.aic)
            SARIMAX_order.append([trend_param, seasonal_param])
            
        except : 
            continue
      
        
```

```python
# 최상의 파라미터 조합찾기

best_order = SARIMAX_order[AIC.index(min(AIC))]

print(best_order," " ,min(AIC))
```

```python
# best 조합으로 모델학습 시키기 
best_model = sm.tsa.SARIMAX(endog = y_train, exog = x_train, 
                                    order = best_order[0], 
                                                       seasonal_order = best_order[1])
model_fit = best_model.fit(disp = False, maxiter = 250)                                                                                        
```

### Step3. 모형 구조 확인


```python
model_fit.summary()
```

```python
# 잔차확인
model_fit.plot_diagnostics(figsize=(11, 8))
plt.show()


# 일부의 잔차가 이상치로 확인
# KDE가 정규분포를 따르고 있지않음
# 잔차의 분포가 양쪽 끝에서 정규분포를 벗어나는 것을 확인할 수 있음 
# 잔차의 자기상관성이 보이 않음으로 기존의 계절성을 모델이 효과적으로 해결함.
```

### Step 4. 차량 수요 예측

```python
# RMSE
predict = mod_fit.get_forecast(steps = 52, exog = x_test)
predict_mean = predict.predicted_mean
model_rmse = np.sqrt(sklearn.metrics.mean_squared_error(y_test , predict_mean))
print(model_rmse)
```

```python
# 시각화
fig, ax = plt.subplots(figsize=(9,4))
plt.xticks(rotation = 45)
loc = plticker.MultipleLocator(base = 5.0)
ax.xaxis.set_major_locator(loc)

x_axis = y_test[-50:].index
plt.plot(x_axis, y_test[-50:], label = "real")
plt.plot(x_axis, predict_mean[-50:], label = "predict")
plt.legend()


plt.show()
```

## 문제 해결과정

1. SARIMAX모델의 출력결과를 확인하던 중에 RUNNING THE L-BFGS-B CODE라는 error 발생.
    - "RUNNING THE L-BFGS-B CODE" 메시지는 'statsmodels'의 'SARIMAX'모델이 내부적으로 L-BFGS-B알고리즘을 사용하여 최적화를 수행하고 있다는 것을 나타낸다. 해당 메세지는 사용자에게 현재 최적화 알고리즘을 사용해서 모델 파라미터를 조정하고 있는 중임을 알려주는 것이지만 출력값이 너무 길어지기 때문에 disp = False로 해주면 메세지가 나오지 않는다. 
    - 'disp = False' : 모델 피팅 결과를 출력할지 여부를 결정한다. False로 설정하면 최적화 결과가 출력되지 않는다. 
    
2. get_forecast함수 사용중 ValueError: Out-of-sample operations in a model with a regression component require additional exogenous values via the `exog` argument라는 error 발생.
    - 이 error는 'SARIMAX' 모델에서 회귀구성요소가 포함된 경우, out-of-sample(외삽)작업을 수행할 떄 'exof'매개 변수를 통해 추가적인 외생변수값을 제공해야함을 나타낸다. 
    - mod.get_forecast(steps=n, exog=exog_values)와 같이 해결할수 있음.
    - 위의 방법대로 했는데 ValueError: Provided exogenous values are not of the appropriate shape. Required (10, 17), got (571, 17)가 남 
    - train_data로 예측을 해서 target_data 갯수와 안맞았음.
    so, x_test데이터로 바꾸고 steps에 x_test길이와 맞는 값을 넣어줌
    
3. 랜덤 포레스트모델을 적용하던중 ValueError: Unknown label type: 'continuous'해당 error 발생.
    - 우리는 선형회귀 모델로 예측을 해야한느데 분류 모델로 잘못 import함.
    so, 선형회귀 모델로 다시 import함.
